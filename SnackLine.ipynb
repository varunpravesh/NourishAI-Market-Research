```python
import pandas as pd
import numpy as np
import random
from faker import Faker
```

```python
# Initialize Faker
fake = Faker()

num_customers = 1000  

age_groups = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
income_levels = ["Low", "Medium", "High"]
regions = ["North", "South", "East", "West"]

purchase_frequency = ["Daily", "Weekly", "Monthly", "Rarely", "Never"]

health_preferences = ["Low-calorie", "High-protein", "Organic", "Gluten-free", "Vegan", "No preference"]
brands = ["NourishAI", "Competitor A", "Competitor B", "Competitor C", "No preference"]

# Free-text responses about snacks
snack_responses = [
    "A quick energy boost during work.",
    "A healthy alternative to junk food.",
    "Something crunchy and tasty to enjoy.",
    "A guilty pleasure for movie nights.",
    "A way to try new flavors and ingredients."
]


data = {
    "CustomerID": range(1, num_customers + 1),
    "Age Group": np.random.choice(age_groups, num_customers),
    "Income Level": np.random.choice(income_levels, num_customers),
    "Region": np.random.choice(regions, num_customers),
    "Purchase Frequency": np.random.choice(purchase_frequency, num_customers, p=[0.1, 0.3, 0.3, 0.2, 0.1]),
    "Health Preference": np.random.choice(health_preferences, num_customers),
    "Brand Recall": np.random.choice(brands, num_customers, p=[0.4, 0.2, 0.15, 0.15, 0.1]),
    "Snack Perception": np.random.choice(snack_responses, num_customers)
}

df = pd.DataFrame(data)
df.to_csv("SnackLine.csv", index=False)
```
